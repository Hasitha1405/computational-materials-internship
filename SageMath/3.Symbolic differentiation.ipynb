```python
f.diff(x).show()

```

```python
diff(f,x).show()
```

```python
diff(e^sin(x^2),x).show()
```

VISUALIZING

```python
g(x)=sin(x+pi/2)
plot(g(x), -5,5, color='red')
```

```python
dg(x)=g.diff(x)
fig1=plot(g(x), -5,5, color='red', legend_label="sinx")
fig2=plot(dg(x), -5,5, color='blue', legend_label="d(sinx)")
fig=fig1+fig2
fig.show(legend_loc='best')
```

MULTIPLE VARIABLES

```python
p(x,y) = 3*x^4*y^3 + 9*y*x^2 - 4*x + 8*y
p.show()
```

```python
dwx(x,y)=diff(p,x)
dwx.show()
```

```python
dwy(x,y)=diff(p,y)
dwy.show()
```

```python
d21(x,y)=diff(p,x,y)
d22(x,y)=diff(p,y,x)

d21.show()
d22.show()
```

```python
dx2=diff(p,x,2)
dx2.show()
```
